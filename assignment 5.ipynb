#Q4
When conducting a hypothesis test, we first formulate the “null hypothesis,” (the assumption that there are no effects or differences in the population). We then collect samples, calculate the test statistic, and compare it to the expected result if the hypothesis is true.

The p-value is the probability of obtaining a test statistic that is as extreme as the observed statistic when the null hypothesis is assumed to be true. It tells us how likely we are to get the data if the null hypothesis is correct.

A smaller p-value means that the probability of seeing the results we observed or more extreme results if the null hypothesis is true is very low. the smaller the p-value, the stronger the evidence against the null hypothesis, making the null hypothesis seem more difficult to realize.
A smaller p-value indicates that the observed test statistic is far from what we would expect under the null hypothesis, making us skeptical of the null hypothesis. If the null hypothesis is true, we are less likely to see the results we observe.


#Q8

Sample Size: Our experiment uses a much larger sample (80 students) compared to Fisher's original 8 trials with Dr. Bristol.
Population of Interest: While Fisher's experiment focused on Dr. Bristol's specific ability, our study examines a more general population of STA130 students.
Parameter of Interest: In our case, we're interested in the proportion of STA130 students who can correctly identify the pouring order, rather than an individual's ability.
Null Hypothesis (H₀):

Formal: p = 0.5
Informal: STA130 students are guessing randomly (no better than chance) when identifying whether milk or tea was poured first.

Alternative Hypothesis (H₁):

Formal: p > 0.5
Informal: STA130 students can identify the pouring order better than random guessing would suggest.

I will use a hypothesis testing approach, specifically a one-sided binomial test. 


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(130)

n = 80  
x = 49  

n_simulations = 100000

simulated_results = np.random.binomial(n, 0.5, n_simulations)

p_value = np.mean(simulated_results >= x)

plt.figure(figsize=(10, 6))
plt.hist(simulated_results, bins=30, density=True, alpha=0.7)
plt.axvline(x, color='red', linestyle='dashed', linewidth=2)
plt.title('Simulated Distribution under Null Hypothesis')
plt.xlabel('Number of Correct Identifications')
plt.ylabel('Density')
plt.annotate(f'Observed: {x}', xy=(x, 0), xytext=(x+1, 0.01),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()

print(f"Estimated p-value: {p_value:.4f}")
```

Supporting Visualization
The histogram shows the simulated distribution of correct identifications under the null hypothesis, with the observed value marked in red.

Discussion
Conclusion Regarding the Null Hypothesis
Based on our simulation, we estimated a p-value of approximately {p_value:.4f}. This p-value represents the probability of observing 49 or more correct identifications out of 80 trials if the null hypothesis (random guessing) were true.

If we use a traditional significance level of 0.05, we would [reject/fail to reject] the null hypothesis, as our p-value is [less than/greater than] 0.05.
This suggests [strong/moderate/weak] evidence against the null hypothesis of random guessing.
We can conclude that there is [significant/some/little] evidence to support the claim that STA130 students can identify the pouring order better than chance.



```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(130)

# Observed data
n = 80  # sample size
x = 49  # number of correct identifications

# Simulation parameters
n_simulations = 100000

# Simulate under null hypothesis
simulated_results = np.random.binomial(n, 0.5, n_simulations)

# Calculate p-value
p_value = np.mean(simulated_results >= x)

# Visualize the simulation results
plt.figure(figsize=(10, 6))
plt.hist(simulated_results, bins=30, density=True, alpha=0.7)
plt.axvline(x, color='red', linestyle='dashed', linewidth=2)
plt.title('Simulated Distribution under Null Hypothesis')
plt.xlabel('Number of Correct Identifications')
plt.ylabel('Density')
plt.annotate(f'Observed: {x}', xy=(x, 0), xytext=(x+1, 0.01),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()

print(f"Estimated p-value: {p_value:.4f}")
```
